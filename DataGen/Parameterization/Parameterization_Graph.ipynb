```python
import numpy as np
import plotly.graph_objs as go
from ipywidgets import FloatSlider, VBox, HBox

# 定义表达式
def expression(x, y, z, a, b, c):
    return np.log(a * x + b ** np.log(y) + c **(10* z))

# 设置初始参数值
a_init = 1
b_init = 2
c_init = 3
y_init = 2
z_init = 3

# 定义x的范围
x = np.linspace(0, 1000, 10)

# 创建初始图表
T = expression(x, y_init, z_init, a_init, b_init, c_init)
line = go.Scatter(x=x, y=T, mode='lines', line=dict(color='blue'))

layout = go.Layout(
    xaxis_title='x',
    yaxis_title='t',
    width=800,  # 设置图表宽度
    height=600  # 设置图表高度
)

fig = go.FigureWidget(data=[line], layout=layout)

# 创建交互式滑动条
a_slider = FloatSlider(value=a_init, min=0, max=10.0, step=0.1, description='a:')
b_slider = FloatSlider(value=b_init, min=0, max=10.0, step=0.1, description='b:')
c_slider = FloatSlider(value=c_init, min=0, max=10.0, step=0.1, description='c:')
y_slider = FloatSlider(value=y_init, min=0, max=10000.0, step=100, description='y:')
z_slider = FloatSlider(value=z_init, min=0, max=1, step=0.01, description='z:')

# 更新图表函数
def update_plot(change):
    T = expression(x, y_slider.value, z_slider.value, a_slider.value, b_slider.value, c_slider.value)
    with fig.batch_update():
        fig.data[0].y = T

# 为滑动条绑定更新函数
a_slider.observe(update_plot, names='value')
b_slider.observe(update_plot, names='value')
c_slider.observe(update_plot, names='value')
y_slider.observe(update_plot, names='value')
z_slider.observe(update_plot, names='value')

# 显示滑动条和图表
display(VBox([fig, HBox([a_slider, b_slider, c_slider]), HBox([y_slider, z_slider])]))
```
